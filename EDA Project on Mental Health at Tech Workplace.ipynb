# **1. Introduction**
---

- Mental health is a crucial aspect of our overall well-being, influencing how we think, feel, and interact with the world around us. It encompasses our emotional resilience, our ability to handle stress, and our capacity to form and maintain healthy relationships. Good mental health is vital for leading a fulfilling and productive life. 

- Work constitutes a significant portion of our lives, where we invest a substantial amount of our time to earn income and cultivate friendships. Having a satisfying job can positively impact our mental well-being and overall quality of life.


 
- Understanding the importance of mental health is essential because it affects every facet of our daily existence. Here are some key points to consider:

  - **Well-Being:** Good mental health contributes to a sense of well-being, allowing individuals to experience life's joys and challenges more effectively.
  - **Productivity:** Mental health plays a significant role in our productivity, both at work and in our personal lives. It influences our motivation, focus, and problem-solving abilities.  
  - **Relationships:** Healthy mental well-being fosters positive relationships. It enables us to communicate effectively, empathize with others, and build strong connections.  
  - **Physical Health:** Mental health and physical health are interconnected. Neglecting one can affect the other, leading to various health issues.  
  - **Resilience:** Good mental health equips individuals with the tools to cope with life's ups and downs. It helps us bounce back from adversity and adapt to change.

- Addressing mental health issues requires a multi-faceted approach, including raising awareness, reducing stigma, providing access to mental health services, and promoting self-care and emotional well-being. It's essential for individuals to prioritize their mental health and seek help when needed, just as they would for physical health concerns.
- Recognizing its importance and addressing mental health issues at home as well as at work place are crucial steps toward a healthier and more resilient society.

<a name = Section61></a>
### **6.1 Handling of Missing Data**

- In this section, we will identify missing data and check the proportion of it and take measures accordingly.

```python
# identifying the sum of missing values in all the features
nf = pd.DataFrame(index = mh_data.columns.values)
nf['Null Frequency'] = mh_data.isnull().sum().values

# calculating the % of missing values
percent = mh_data.isnull().sum().values/mh_data.shape[0]
nf['Missing %age'] = np.round(percent, decimals = 4) * 100
nf.transpose()
```

**Observations:**

- We can observe that following features are found to have missing values along with the proportions:

|Feature|Object Type|Missing Proportion|Solution|
|:--:|:--:|:--:|:--|
|state|Object|40%|(Only US State Captured) Rest will fill with Non USA |
|self_employed|Object|1.43%|Replace with mode|
|work_interfere|Object|20.97%|Replace with mode|
|comments|Object|86.97%|Drop the feature|

```python
# replacing NaN values present in 'self_employed' with mode
mh_data['self_employed'] = mh_data['self_employed'].replace(np.nan, mh_data['self_employed'].mode()[0])

# replacing NaN values present in 'work_interfere' with mode
mh_data['work_interfere'] = mh_data['work_interfere'].replace(np.nan, mh_data['work_interfere'].mode()[0])

#filling NaN values present in 'state' with 'Non USA'
mh_data['state'] = mh_data['state'].fillna('Non USA')

# dropping 'comments' using pandas drop method
mh_data.drop(labels='comments', axis=1, inplace=True)
```

```python
# verifying if the changes have been effectively done or not
nf = pd.DataFrame(index = mh_data.columns.values)
nf['Null Frequency'] = mh_data.isnull().sum().values

# calculating the % of missing values
percent = mh_data.isnull().sum().values/mh_data.shape[0]
nf['Missing %age'] = np.round(percent, decimals = 4) * 100
nf.transpose()
```

### **6.3 Handling of Inconsistent Data**

```python
# transforming Timestamp's object datatype to datetime
mh_data['timestamp'] = pd.to_datetime(mh_data['timestamp'])
```

```python
# changing the different formats, if any, to one
mh_data['timestamp'] = pd.to_datetime(arg = mh_data['timestamp'], format = '%d-%m-%Y')
```

```python
# segregating and aligning all these responses into 3 categories: male, female and other
# categorizing 'male' for the resposes seems to be coming user this category
male = ['M','Male', 'male', 'm', 'maile', 'Cis Male', 'Mal', 'Male (CIS)', 'Make', 'Male ', 
        'Man', 'Mail', 'cis male', 'Malr', 'Cis Man']

# categorizing 'female' for the resposes seems to be coming user this category
female = ['Female', 'female', 'Cis Female', 'F', 'Woman', 'f', 'Femake', 'woman', 'Female ', 
          'cis-female/femme', 'msle', 'Female (cis)', 'femail']
```

```python
# replacing as categorized above
mh_data['gender'] = mh_data['gender'].replace(male, 'male').replace(female, 'female')
```

```python
# any other responses have been categorized as 'other'
filter_mask = mh_data['gender'].isin(['male', 'female'])
mh_data.loc[~filter_mask, 'gender'] = 'other'
```

```python
# verifying if the changes have been effectively done or not
mh_data['gender'].value_counts()
```

<a name = Section64></a>
### **6.4 Handling of Incorrect Values**

- **age** feature includes some **logically incorrect numbers** like 999999999999, -1729.

- These are **incorrect values** and we will **perform capping** over these values such as:
  - All value above 75 will be capped to 75 (on **average 65** is the **retirement** but ensuring extra buffer).
  - All value below 14 will be capped to 14 (on **average 14** is the **minimum age** for **employment**).

- As a general rule, the FLSA sets 14 years old as the minimum age for employment under restricted number of hours.

```python
# using the clip function to replace incorrect age values
mh_data['age'] = mh_data['age'].clip(lower = 14, upper = 75)
```

- The correlation is drawn with **Treatment** and the other variables and we see, it is **correlated** with the following:
  - **Family History**, **Care Options**, **Benefits** and **Obs_Consequence**
